```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

appended_data['days_squared'] = appended_data['days'] ** 2

# Create an interaction term between 'treated' and 'post' to capture the DiD estimate
# This term will estimate the differential effect of the treatment in the post-period
appended_data['treatment_interaction'] = appended_data['treated'] * appended_data['post']

appended_data['treated_days'] = appended_data['treated'] * appended_data['days']
appended_data['treated_days_squared'] = appended_data['treated'] * appended_data['days_squared']
policy_start_date = '2019-11-4'
appended_data['days_since_policy'] = (appended_data['date'] - pd.to_datetime(policy_start_date)).dt.days

appended_data['post_days'] = appended_data['post'] * appended_data['days_since_policy']
# appended_data['post_month_squared'] = appended_data['post'] * appended_data['month_squared']
appended_data['treated_post_policy_time'] = appended_data['treated'] * appended_data['post'] * appended_data['days_since_policy']
appended_data['pm25_lagged'] = appended_data['pm25'].shift(1)
appended_data['post_sun'] = appended_data['post_sun']*appended_data['treated']
appended_data = appended_data.dropna(subset=['pm25_lagged'])

did_model = smf.ols(
    'pm25 ~ treated + post + days + treatment_interaction + diwali + pm25_lagged + post_sun',
    data=appended_data
).fit(cov_type='HC3')

print(did_model.summary())
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming 'appended_data' has 'treated', 'post', 'month', and 'pm25' variables

# Calculate the squared month term
appended_data['days_squared'] = appended_data['days'] ** 2

# Create interaction terms between 'treated' and 'post'
appended_data['treatment_interaction'] = appended_data['treated'] * appended_data['post']

# Create interaction terms between 'treated' and 'month' variables
appended_data['treated_days'] = appended_data['treated'] * appended_data['days']
appended_data['treated_days_squared'] = appended_data['treated'] * appended_data['days_squared']

# Create interaction terms between 'post' and 'month' variables to capture changing policy effects over time
policy_start_date = '2023-10-12'
appended_data['days_since_policy'] = (appended_data['date'] - pd.to_datetime(policy_start_date)).dt.days
appended_data['treated_post_policy_time'] = appended_data['treated'] * appended_data['post'] * appended_data['days_since_policy']
# Define the DiD model with added interaction terms to capture evolving policy effects
did_model = smf.ols(
    'pm25 ~ treated + post + days + days_squared + treated_days + treated_days_squared + treatment_interaction + diwali',
    data=appended_data
).fit(cov_type='HC3')  # Use heteroskedasticity-robust standard errors

# Print the summary of the regression with robust standard errors
print(did_model.summary())
```

```python
# Assuming 'policy_day' is a variable representing the day since the policy started (0 for the first day, 1 for the second, etc.)
# appended_data['policy_day'] = (appended_data['date'] - pd.to_datetime(policy_start_date)).dt.days

# Interaction term for each day in the treatment period
appended_data['treated_policy_day'] = appended_data['treated'] * appended_data['days_since_policy']

# Define the daily DiD model
daily_did_model = smf.ols(
    'pm25 ~ treated + days_since_policy + treated_policy_day',
    data=appended_data[appended_data['post'] == 1]  # Limit to the 7-day policy period
).fit(cov_type='HC3')

# Print the summary
print(daily_did_model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Extract the design matrix (X) from the model
X = appended_data[['treated', 'post', 'days', 'days_squared','treatment_interaction', 'diwali', 'pm25_lagged', 'post_sun']]
# Calculate VIF for each predictortreated + post + days + treatment_interaction + diwali + pm25_lagged + post_sun',
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
# treated + post + days + days_squared + treated_days + treated_days_squared + treatment_interaction + treated_post_policy_time
print(vif_data)
```

```python
# Assuming 'appended_data' is your DataFrame and it has the 'pm25', 'treated', and 'post' columns
# 'month' is a categorical variable representing the month of observation

# Create month dummies
month_dummies = pd.get_dummies(appended_data['month'], prefix='month', drop_first=True)
appended_data = pd.concat([appended_data, month_dummies], axis=1)

# Create interaction terms between 'treated' and month dummies
for month in month_dummies.columns:
    appended_data[f'treated_{month}'] = appended_data['treated'] * appended_data[month]

# Define the model formula with month dummies and their interaction with 'treated'
formula = 'pm25 ~ treated + post'
# Add month dummies and their interactions to the formula
for month in month_dummies.columns:
    formula += f' + {month} + treated_{month}'
# Add the treatment interaction term
formula += ' + treatment_interaction'

# Define the DiD model with month dummies and interaction terms
did_model = smf.ols(formula, data=appended_data).fit()

# Print the summary of the regression
print(did_model.summary())
```
